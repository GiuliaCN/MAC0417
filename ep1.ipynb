### Leitura de imagens

```python
img = io.imread('dataset/tesoura/tesoura-1bdi1.jpg')
```

```python
io.imshow(img)
img.shape
```

```python
imgs = io.imread_collection("dataset/tesoura/*.jpg") # coloque a sua ext
```

```python
io.imshow(imgs[0])
```

```python
n_colunas, n_linhas = 10, 10

fig, axs = plt.subplots(n_linhas, n_colunas, figsize=(20, 20))
num_fotos = n_colunas * n_linhas
for e, objeto in enumerate(imgs[:num_fotos]):
    l, c = (e // n_colunas, e % n_colunas)
    axs[l, c].imshow(objeto)
    axs[l, c].axis('off')
plt.show()
```
